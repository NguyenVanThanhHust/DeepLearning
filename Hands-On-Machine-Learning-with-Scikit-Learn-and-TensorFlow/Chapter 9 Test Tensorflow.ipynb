```python
import tensorflow as tf
x = tf.Variable(3, name = "x")
y = tf.Variable(5, name = "y")
f = x*x*y +y+2
```

```python
sess = tf.Session()
sess.run(x.initializer)
sess.run(y.initializer)
result = sess.run(f)
```

```python
f
```

```python
result
```

```python
with tf.Session() as sess:
    x.initializer.run()
    y.initializer.run()
    result = f.eval()
```

```python
result
```

```python
sess.close()
```

```python
result
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    init.run()
    result = f.eval()
```

```python
result
```

```python
sess.close()
```
