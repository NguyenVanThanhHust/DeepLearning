For f(x) = ax

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
sess = tf.Session()
```

```python
#declare our model variable ,placeholder
a = tf.Variable(tf.constant(4.0))
x_val = 5.0
x_data = tf.placeholder(dtype = tf.float32)
```

```python
# Operation for computational graph
multiplication = tf.multiply(a,x_data)
#Loss function
loss = tf.square(tf.subtract(multiplication, 50.0))
```

```python
#initialize our model and declare our optimizing algorith
init = tf.global_variables_initializer()
sess.run(init)
my_opt = tf.train.GradientDescentOptimizer(0.01)
train_step = my_opt.minimize(loss)
```

```python
print('Optimizing a Multiplication Gate Output to 50.0')
for i in range(15):
    sess.run(train_step, feed_dict={x_data:x_val})
    a_val = sess.run(a)
    mult_output = sess.run(multiplication, feed_dict={x_data:x_val})
    print(str(a_val) + ' * ' + str(x_val) + ' = ' + str(mult_output))
```

For f(x) = ax+b

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
sess = tf.Session()
```

```python
#declare our model variable ,placeholder
a = tf.Variable(tf.constant(4.0))
b = tf.Variable(tf.constant(5.0))
x_val = 5.0
x_data = tf.placeholder(dtype = tf.float32)
```

```python
# Operation for computational graph
multiplication = tf.multiply(a,x_data)
addition = tf.add(b, multiplication)
#Loss function
loss = tf.square(tf.subtract(addition, 50.0))
```

```python
#initialize our model and declare our optimizing algorith
init = tf.global_variables_initializer()
sess.run(init)
my_opt = tf.train.GradientDescentOptimizer(0.01)
train_step = my_opt.minimize(loss)
```

```python
print('Optimizing a Multiplication Gate Output to 50.0')
for i in range(15):
    sess.run(train_step, feed_dict={x_data:x_val})
    a_val = sess.run(a)
    b_val = sess.run(b)
    addition_output = sess.run(addition, feed_dict={x_data:x_val})
    print(str(a_val) + ' * ' + str(x_val) + str(b_val) + ' = ' + str(addition_output))
```
